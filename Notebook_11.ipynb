```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
df = pd.read_excel('data/Niemeyer_2016/phosphines_parameterization.xlsx', index_col=0)
df
```

```python
import seaborn as sns
```

```python
fig, ax = plt.subplots(figsize=(7,6))
sns.heatmap(df.corr(), ax=ax)
plt.show()
```

```python
def plot_(ax, x, y, df):
    ax.scatter(df[x], df[y], c=df["DDG_activation"], alpha=0.8)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    
fig, ax = plt.subplots(2, 2, figsize=(6,6))

plot_(ax[0,0], "P_31P NMR shift", "P-C bend max", df)
plot_(ax[0,1], "P_31P NMR shift", "PSe_31P NMR shift", df)
plot_(ax[1,0], "P_31P NMR shift", "Cone min", df)
plot_(ax[1,1], "P_31P NMR shift", "P=O max", df)
```
