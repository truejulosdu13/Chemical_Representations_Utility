```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams['text.usetex'] = True
```

```python
df = pd.read_excel("data/Wu_Doyle/phosphines_yields_descriptos.xlsx")
```

```python
df
```

```python
def plot_(ax, x, y, df):
    ax.scatter(df[x], df[y], c=df["Yield"], alpha=0.8)
    ax.set_xticks([])
    ax.set_yticks([])
    ax.set_xlabel(x)
    ax.set_ylabel(y)
    
fig, ax = plt.subplots(2, 2, figsize=(6,6))

plot_(ax[0,0], "%Vbur", "Vmin", df)
plot_(ax[0,1], "%Vbur", "ConeAngle", df)
plot_(ax[1,0], "ConeAngle", "Vmin", df)
```
